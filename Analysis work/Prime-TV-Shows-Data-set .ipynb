```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Ecom = pd.read_excel('F:\Skillenable\Data frames\E Commerce Analysis - Assignment Data set.xlsx')
Ecom
```

```python
#1) Create new data frame by grouping the data of average quantity based on product line.

avgQuality_prodLine = Ecom.groupby('Product line')[['Quantity']].mean()
avgQuality_prodLine
```

```python
#2) Find the number of orders from customers who are Male and bought Sports and travel product line.

fc1 = Ecom['Gender'] == 'Male'
fc2 = Ecom['Product line'] == 'Sports and travel'

Ecom[fc1 & fc2].shape[0]

#Conclusion:  There are 78 orders from customers who are Male and bought Sports and travel product line.
```

```python
#3) Find the top 4 invoice id based on the cogs column.

top4_cogs_invoice = Ecom.sort_values(by = 'cogs', ascending= False).head(4)
top4_cogs_invoice['Invoice ID']


```

```python
#4) Find out the product line which has maximum sum of unit price.

max_unit_price = Ecom.groupby('Product line')[['Unit price']].sum().sort_values(by='Unit price', ascending = False)
max_unit_price.head(1)

# Fashion accessories has maximum sum of unit price.
```

```python
#5) Find the average cogs of each city.

AvgCogs_city = Ecom.groupby('City')[['cogs']].mean()
AvgCogs_city
```

```python
#6) From the Date column, find the day name for each order. Find the top 3 days based on the average rating.

from datetime import date as dt 
```

```python
Ecom['Day_name']= Ecom['Date'].dt.day_name()

top3_days = Ecom.groupby('Day_name')[['Rating']].mean().sort_values(by = "Rating", ascending = False)
top3_days.head(3)

# Monday, friday, Sunday are the top 3 days based on the average rating
```

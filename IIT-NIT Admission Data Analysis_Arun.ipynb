```python
# converting series data to dataframe
df_adm = max_adm_quota.to_frame()
df_adm
```

```python
#resetting index
df_adm.reset_index(inplace=True)
df_adm
```

```python
#renaming columns

df_adm.columns =['Categories', 'No_of_admission']
df_adm
  
```

```python
# converting to list
Categories = df_adm['Categories'].tolist()
No_of_admission = df_adm['No_of_admission'].tolist()

print(Categories)
print(No_of_admission)
```

```python
# creating the bar plot

fig = plt.figure(figsize = (10, 5))
 

plt.bar(Categories, No_of_admission, color ='blue',
        width = 0.4)
 
plt.xlabel("Categories")
plt.ylabel("No. of Admission")
plt.title("Category vise admissions in overall years.")
plt.show()
```

```python
adm_accepted = data[data.is_preparatory==1]
adm_rejected = data[data.is_preparatory==0]
```

## What is the most optimum Opening and closing rank?

```python
# Opening rank
#Minimum and Maximum 

min_opening_rank = data['opening_rank'].min()
max_opening_rank = data['opening_rank'].max()
print("The Minimum Opening Rank is ", min_opening_rank, " and Maximum Opening Rank is- ", max_opening_rank)
```

```python
# Average ranking

avg_opening_rank = round(data['opening_rank'].mean(axis = 0))

print("Avergae opening rank over the years has been - ", avg_opening_rank)
```
